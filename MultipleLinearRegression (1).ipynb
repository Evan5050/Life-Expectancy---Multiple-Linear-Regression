```python
predictions_df = pd.DataFrame({"Status": df["Status"],
                          "HIV/AIDS": df["HIV/AIDS"],
                          "Income Composition Of Resources": df["Incomecompositionofresources"],
                          "Schooling": df["Schooling"],
                          "Life Expectancy": df["life"],
                          "Sklearn Predictions": sklearn_y_predictions})
predictions_df
#based on all of the 4 inputs it gives prediction
```

```python
# Assuming lifeexpectancy[i] = alpha + (beta_1 * HIV/AIDS[i]) + (beta_2 * Income Composition Of Resources[i]) + (beta_3 * Schooling[i]) + (beta_4 * Status[i]) + error
# Generally: y[i] = alpha + (beta_1 * x_1[i]) + (beta_2 * x_2[i]) + (beta_3 * x_3[i]) + (beta_4 * x_4[i]) + error 
# Model:     y_hat[i] = alpha_hat + (beta_1_hat * x_1[i]) + (beta_2_hat * x_2[i]) + (beta_3_hat * x_3[i]) + (beta_4_hat * x_4[i])

def get_predictions(model, x):
    '''
    Returns the predictions for the given model and inputs

    Input: model - a np.array of floats with the shape of (p,) of parameters
           x - np.array of floats with the shape of (n, p-1) of inputs
  
    Output: returns a np.array of floats with a shape of (n,)
    '''
    (n, p_minus_one) = x.shape 
    p = p_minus_one + 1

    new_x = np.ones(shape=(n,p)) #n by p array of ones
    new_x[:, 1:] = x #everything but first collum is overridden

    return np.dot(new_x, model)
```

```python
test_model = np.array([1,2,1,4,7]) #random arbitrary test model
get_predictions(test_model,x_train).shape
```

```python
predictions_df["test predictions"] = get_predictions(test_model,x_train)
predictions_df
```

```python
mean_absolute_error(predictions_df["test predictions"], y_train)
```

```python
from numpy.linalg import inv


def get_best_model(x,y):
    '''
    Returns the model that minimizes the MSE with the given parameters

    Input: x - a np.array of floats with the shape (n, p-1) of inputs
           y - a np.array of floats with the shape (n,) of observed outputs

    Output: returns a np.array with the shape of (p,) representing the model
    '''
    (n,p_minus_one) = x.shape
    p = p_minus_one + 1

    new_X = np.ones(shape=(n, p))
    new_X[:, 1:] = x

    return np.dot(np.dot(inv(np.dot(new_X.T, new_X)), new_X.T), y)
    #emulates fit model
```

```python
best_model = get_best_model(x_train, y_train)
predictions_df["Best Predictions"] = get_predictions(best_model, x_train)
predictions_df
```

```python
mean_absolute_error(predictions_df['Best Predictions'], y_train) , mean_squared_error(predictions_df['Best Predictions'], y_train)
```

```python
np.set_printoptions(legacy='1.25')

def mean_squared_error(y, predictions):
  '''
  Returns the mean squared error given the observed and predicted outputs.
  
  Input: y - np.array of floats with the shape (n,) of observed outputs
         y_predictions - np.array of floats with the shape (n,) of predicted outputs

  Output: returns afloat representing the MSE.
  '''
  n = len(y)
  return np.sum(np.square(y - predictions)) / n
  #sum of squared distances of y - y_predictions(errors/difference between them) all divided by n to get mean

def mean_absolute_error(y, predictions):
  '''
  Returns the mean absolute error given the observed and predicted outputs.
  
  Input: y - np.array of floats with the shape (n,) of observed outputs
         y_predictions - np.array of floats with the shape (n,) of predicted outputs

  Output: returns float representing the MSE.
  '''
  n = len(y)
  return np.sum(np.abs(y - predictions)) / n
  #same but the sum of the absolutes

mean_absolute_error(predictions_df['Best Predictions'], y_train) , mean_squared_error(predictions_df['Best Predictions'], y_train)
```
